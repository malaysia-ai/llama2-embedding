```python
from glob import glob
from tqdm import tqdm
import json
```

```python
files = glob('iium/*.jsonl')
len(files)
```

```python
with open('/home/ubuntu/embedding-pair-mining/iium-train-dataset-bge.jsonl', 'w') as fopen_l:
    for f in files:
        with open(f) as fopen:
            for l in tqdm(fopen):
                l = json.loads(l)
                if not len(l['pos']) or not len(l['neg']):
                    continue
                fopen_l.write(f'{json.dumps(l)}\n')
                fopen_l.flush()
```
